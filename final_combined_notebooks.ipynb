```python
# read in all csvs
d1 = "resources/10308_fuel_use_veh_type_3-25-20.csv"
d2 = "resources/epa_transport_data.csv"
d3 = "resources/table_01_11_053019.csv"
d4 = "resources/table_01_19_120419.csv"
```

```python
# df1 first round cleanup
del df1["Source"]
new_names1 = ['vehicle_type',
'mpg_gasoline',
'mpg_diesel',
'vmt',
'annual_fuel_use_gge']
df1.columns = new_names1
df1['vmt'] = df1.vmt.replace(',','', regex=True).astype('int64')
df1['annual_fuel_use_gge'] = df1.annual_fuel_use_gge.replace(',','', regex=True).astype('int64')
df1
```

```python
# df2 first round cleanup
years = []
for i in range(0,7):
    year=(2000+i)
    years.append(str(year))
df2 = df2.rename(columns={"Transportation Sector":"Type"})
df2 = df2.set_index("Type")
df2 = df2[years]
df2 = df2.drop("Total")
df2 = df2.transpose()
df2 = df2.reset_index()
new_names2 = ['year',
'fossil_fuel_combustion_carbon_dioxide',
'use_of_fluorinated_gases',
'fossil_fuel_combustion_other_greenhouse_gases',
'other_transportation_categories']
df2.columns = new_names2
df2['year'] = df2.year.astype('int64')
df2
```

```python
# df3 first round cleanup
years = []
for i in range(0,7):
    year=(2000+i)
    years.append(str(year))
df3 = df3.rename(columns={"Unnamed: 0":"Type"})
df3 = df3.set_index("Type")
df3 = df3[years]
df3 = df3.dropna(how="any")
df3 = df3.drop(['Light duty vehicle, short wheel basec', 'Light duty vehicle, long wheel basec'])
df3 = df3.transpose()
df3 = df3.reset_index()
new_names3 = ['year','air_carrier',
 'general_aviation_active_fleet',
 'highway_total_registered_vehicles',
 'passenger_cars',
 'motorcycle',
 'other_2-axle_4-tire_vehicles',
 'truck_single-unit_2-axle_6-tire_or_more',
 'truck_combinationd',
 'bus',
 'motor_bus',
 'light_rail_cars',
 'heavy_rail_cars',
 'trolley_bus',
 'commuter_rail_cars_and_locomotives',
 'demand_responsive',
 'other',
 'class_I_freight_cars',
 'class_I_locomotive',
 'nonclass_I_freight_cars',
 'car_companies_and_shippers_freight_cars',
 'amtrak_passenger_train_car',
 'amtrak_locomotive',
 'nonself-propelled_vessels',
 'self-propelled_vessels',
 'oceangoing_self-propelled_vessels_1000_gross_tons_and_over',
 'recreational_boats']
df3.columns = new_names3
df3 = df3.replace(',','', regex=True).astype('int64')
df3
```

```python
# df4 first round cleanup
df4 = df4.rename(columns={"Unnamed: 0":"Type","(R) 2000":"2000","(R) 2001":"2001","(R) 2002":"2002","(R) 2003":"2003","(R) 2004":"2004","(R) 2005":"2005","(R) 2006":"2006","(R) 2007":"2007","(R) 2008":"2008","(R) 2009":"2009","(R) 2010":"2010","(R) 2011":"2011","(R) 2012":"2012","(R) 2013":"2013","(R) 2014":"2014","(R) 2015":"2015","(R) 2016":"2016","(R) 2017":"2017"})
df4 = df4.set_index("Type")
df4 = df4.replace("Z",0)
df4 = df4.transpose()
df4 = df4.reset_index()
new_names4 = ['year',
'hybrid_electric',
'plug-in_hybrid-electric',
'electric']
df4.columns = new_names4
df4 = df4.replace(',','', regex=True).astype('int64')
df4
```

```python
# Maximizes window, this will help deal with errors for all clicking functions
browser.driver.maximize_window()
```

```python
# visiting database
url = 'http://ghgdata.epa.gov/ghgp/main.do#'
browser.visit(url)
time.sleep(1)
```

```python
# Find by id (of a div tag). Within that div tag, the img tag is the "x" of the pop up.
button = browser.find_by_id('welcomeWindowEmitters').find_by_tag('img')
time.sleep(1)
```

```python
# clicks on the img that was defined by using "button" variable
button.click()
```

```python
# activates beautiful soup and parses with html
html = browser.html
soup = BeautifulSoup(html, 'html.parser')
#finds the tags where the states in the dropdown menu are
choose_state = soup.find("select", id="parentState").find_all("option")
```

```python
# takes just the first 4 states
all_states = choose_state[1:5]
df_list = []
```

```python
# for every state in the list of states, select that state and click "apply search"
for item in all_states:
    print(item["value"])
    button3 = browser.find_by_id('stateSelection').find_by_id('parentState').find_by_value(item["value"])
    time.sleep(1)
    button3.click()
    time.sleep(1)
    button4 = browser.find_by_id('applySearch')
    time.sleep(1)
    button4.click()
    time.sleep(1)
# if there is a pop up, exit out of it
    try:
        pop_up_close = browser.find_by_id("fsrFocusFirst")
        pop_up_close.click()
# if there isn't a pop up, print "no pop up"
    except:
        print("no pop up")
    time.sleep(1)
# find and click the button with the id "vList"
    button_list = browser.find_by_id('dataView').find_by_id('vList')
    time.sleep(1)
    button_list.click()
    time.sleep(1)
# parse with html
    html = browser.html
    soup = BeautifulSoup(html,"html.parser")
# find and click the button with the text "Total Reported Emissions"
    time.sleep(1)
    button_sort_total_e=browser.find_by_text('Total Reported Emissions')
    time.sleep(1)

    # first click, ascending
    time.sleep(1)
    button_sort_total_e.click()
    time.sleep(1)

    # second click, descending
    time.sleep(1)
    button_sort_total_e.click()
    time.sleep(1)

# parse with html
    html = browser.html
    soup = BeautifulSoup(html,"html.parser")
    
    #read the table
    table=soup.find("div",class_="grid-canvas")
    #convert to dataframe
    df = pd.DataFrame(table)
    df = df.rename(columns = {1:"Facility",2:"County",3:"State",4:"Total Reported Emissions",5:"Sector"})
    df_list.append(df)
```

```python
# creates new empty dataframes
new_sector = pd.DataFrame()
emissions_facilities = pd.DataFrame()

# for every dataframe in the list of dataframe from the previous cell
for x in range(len(df_list)):
# for every column of every dataframe
    for column in df_list[x].columns:
        new_list =[]
# for every row in every column of the current dataframe
        for item in df_list[x][column]:
# add the text of the item (which is the data we need) to a list
            new_list.append(item.text)
# adds the whole list to the column of the dataframe
        new_sector[column] = new_list
# add the column to a dataframe to make one large combined dataframe
    emissions_facilities=emissions_facilities.append(new_sector, ignore_index=True)
    
emissions_facilities
```

```python
# resets the index
emissions_facilities = emissions_facilities.reset_index()
emissions_facilities
```

```python
# deletes unneccessary column
del emissions_facilities[0]
emissions_facilities
```

```python
# deletes column we don't need for project
del emissions_facilities["County"]
emissions_facilities
```

```python
# renames columns so they are in PostgreSQL format
emissions_facilities = emissions_facilities.rename(columns={"index":"id", "Facility":"facility",
                                                           "State":"state", "Total Reported Emissions":"total_reported_emissions",
                                                            "Sector":"sector"})
emissions_facilities
```
