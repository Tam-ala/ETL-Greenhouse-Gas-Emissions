```python
import pandas as pd
from sqlalchemy import create_engine
```

```python
# imported csv file 
electric_file = "resources/world_electricity_gen_source.csv"
electric_data = pd.read_csv(electric_file)
electric_data
```

```python
# renamed columns for SQL use
electric_data = electric_data.rename(columns={"Unnamed: 0":"year", "Coal": "coal", 
                                                    "Oil":"oil", "Natural gas": "natural_gas", "Biofuels":"biofuels", 
                                                    "Waste":"waste", "Nuclear":"nuclear", "Hydro":"water", "Geothermal":"geothermal",
                                                    "Solar PV":"solar_pv", "Solar thermal":"solar_thermal","Wind":"wind",
                                                    "Tide":"tide", "Other sources": "other_sources", "Units":"units"})

electric_data
```

```python
# imported csv file
electric_biofwaste = "resources/world_electricity_gen_from_biofuels_and_waste_by_source.csv"
electric_biofwaste_data = pd.read_csv(electric_biofwaste)
electric_biofwaste_data
```

```python
# renamed columns for SQL use
electric_waste =electric_biofwaste_data.rename(columns={"Unnamed: 0":"year","Industrial waste":"industrial_waste",
                                    "Primary solid biofuels":"primary_solid_biofuels",
                                   "Biogases":"biogases","Municipal waste (renew)":"municipal_waste",
                                   "Liquid biofuels":"liquid_biofuels","Units":"units"})
electric_waste
```

```python
# imported csv file
cait_co2_emiss = "resources/cait_country_co2_emis_energy_subsector.csv"
cait_co2_data = pd.read_csv(cait_co2_emiss)
cait_co2_data.head()
```

```python
# dropped NaNs
wri_data = cait_co2_data.dropna(how="any")
wri_data
```

```python
# renamed columns for SQL use
wri_data =wri_data.rename(columns={"Country":"country","Year":"year",
                                    "Electricity/Heat (CO2) (MtCO2)":"electricity_heat",
                                   "Manufacturing/Construction (CO2) (MtCO2)":"manufactur_construction",
                                    "Transportation (CO2) (MtCO2)":"transportation",
                                    "Other Fuel Combustion (CO2) (MtCO2)":"other_fuel_combust",
                                    "Fugitive Emissions (CO2) (MtCO2)":"fugitive_emissions"})
wri_data
```

```python
# ..........................................................................................
```

```python
# rds_connection_string = "<insert user name>:<insert password>@localhost:5432/customer_db"
rds_connection_string = "postgres:postgres@localhost:5432/emissions_db"
engine = create_engine(f'postgresql://{rds_connection_string}')
```

```python
# check for tables
engine.table_names()
```

```python
# use pandas to load csv convered df into database
electric_data.to_sql(name='electric_data', con=engine, if_exists='append', index=False)
```

```python
# confirm data has been added by doing queries
pd.read_sql_query('select * from electric_data', con=engine).head()
```

```python
# use pandas to load csv convered df into database
electric_waste.to_sql(name='electric_waste', con=engine, if_exists='append', index=False)
```

```python
# confirm data has been added by doing queries
pd.read_sql_query('select * from electric_waste', con=engine).head()
```

```python
# use pandas to load csv convered df into database
wri_data.to_sql(name='wri_data', con=engine, if_exists='append', index=False)
```

```python
# confirm data has been added by doing queries
pd.read_sql_query('select * from wri_data', con=engine).head()
```
